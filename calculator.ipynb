```python
import tkinter as tk

# Create the main application window
root = tk.Tk()
root.title("Calculator")

# Create an entry widget to display calculations
entry = tk.Entry(root, width=16, borderwidth=3, font=('Arial', 24), justify='right')
entry.grid(row=0, column=0, columnspan=4)

# Global variable to store the current expression
current_expression = ""

# Function to update the expression in the entry widget
def button_click(number):
    global current_expression
    current_expression += str(number)
    entry.delete(0, tk.END)
    entry.insert(tk.END, current_expression)

# Function to clear the entry widget
def button_clear():
    global current_expression
    current_expression = ""
    entry.delete(0, tk.END)

# Function to evaluate the expression
def button_equal():
    try:
        global current_expression
        result = str(eval(current_expression))  # Use eval to evaluate the expression
        entry.delete(0, tk.END)
        entry.insert(tk.END, result)
        current_expression = result  # Update current_expression to the result for further calculations
    except:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

# Create number buttons (0-9)
buttons = [
    tk.Button(root, text=str(i), padx=40, pady=20, font=('Arial', 18),
              command=lambda i=i: button_click(i))
    for i in range(10)
]

# Create operation buttons
button_add = tk.Button(root, text="+", padx=38, pady=20, font=('Arial', 18), command=lambda: button_click('+'))
button_subtract = tk.Button(root, text="-", padx=40, pady=20, font=('Arial', 18), command=lambda: button_click('-'))
button_multiply = tk.Button(root, text="*", padx=40, pady=20, font=('Arial', 18), command=lambda: button_click('*'))
button_divide = tk.Button(root, text="/", padx=40, pady=20, font=('Arial', 18), command=lambda: button_click('/'))
button_equal = tk.Button(root, text="=", padx=87, pady=20, font=('Arial', 18), command=button_equal)
button_clear = tk.Button(root, text="C", padx=87, pady=20, font=('Arial', 18), command=button_clear)

# Position number buttons on the grid
buttons[1].grid(row=3, column=0)
buttons[2].grid(row=3, column=1)
buttons[3].grid(row=3, column=2)

buttons[4].grid(row=2, column=0)
buttons[5].grid(row=2, column=1)
buttons[6].grid(row=2, column=2)

buttons[7].grid(row=1, column=0)
buttons[8].grid(row=1, column=1)
buttons[9].grid(row=1, column=2)

buttons[0].grid(row=4, column=0)

# Position operation buttons on the grid
button_add.grid(row=1, column=3)
button_subtract.grid(row=2, column=3)
button_multiply.grid(row=3, column=3)
button_divide.grid(row=4, column=3)

# Position equal and clear buttons
button_equal.grid(row=4, column=1, columnspan=2)
button_clear.grid(row=5, column=0, columnspan=2)

# Start the GUI main loop
root.mainloop()
```
